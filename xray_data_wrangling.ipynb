```python
ncSum = DataParser('data/sci_xrsf-l2-flsum_g15_s20100331_e20200304_v1-0-0.nc')
```

```python
# ncSum.nc.variables.keys()
# flux = ncSum.getVariable('xrsb_flux')
# df = ncSum.createDataFrame(['xrsb_flux', 'background_flux', 'flare_class', 'flare_class'])
# flux[0]
summary = ncSum.getDataFrame()
# summary['flare_class'] = summary['flare_class'].replace('NaN',0)
# summary[:] = np.where(summary.eq('NaN'), 0, summary)
summary['flare_class'].fillna("0",inplace=True)

# One Hot Encode Flare Classes
classDict = {"B": [], "C": [], "M": [], "X": [], "A": []}
for data in summary['flare_class']:
    if data != "0":
        classDict[data[0]].append(1)
        for key in classDict.keys():
            if key != data[0]:
                classDict[key].append(0)
    else:
        for key in classDict.keys():
            classDict[key].append(0)

for key in classDict.keys():
    summary[key] = classDict[key]

flare_types = []
for data in summary['flare_class']:
    flare_types.append(data[0])
summary['flare_type'] = flare_types

summary
```

```python
summary.to_csv("data/xrsSummaryOneHotEncoded.csv")
```
